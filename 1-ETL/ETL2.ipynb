```python
import pandas as pd
```

```python
# Especifica el nombre del archivo Excel
archivo_excel = '../0-DATA/lesiones.xlsx'
```

```python
# Especifica los nombres o índices de las hojas que deseas leer
hojas_a_leer = ['HECHOS', 'VICTIMAS']  # Puedes usar los nombres o los índices de las hojas (0-indexed)
```

```python
# Lee las hojas especificadas en una lista de DataFrames
dataframes = pd.read_excel(archivo_excel, sheet_name=hojas_a_leer)
```

```python
# Ahora puedes acceder a cada DataFrame por su índice o nombre
dataframe_hoja1 = dataframes['HECHOS']
dataframe_hoja2 = dataframes['VICTIMAS']
```

```python
dataframe_hoja1.columns
```

```python
dataframe_hoja2.columns
```

```python
df_merged = pd.merge(dataframe_hoja1, dataframe_hoja2, left_on='id',right_on='ID hecho', how='inner')
#df_merged
```

```python
df_merged.columns
```

```python
df_merged["GRAVEDAD"]=df_merged["GRAVEDAD"].str.upper()
df_merged["gravedad"]=df_merged["gravedad"].str.upper()
```

```python
df_merged['gravedad_final'] = df_merged.apply(lambda row: row['gravedad'] if row['gravedad'] != 'SD' else row['GRAVEDAD'], axis=1)
df_merged
```

```python
df_sin_columnas = df_merged.drop(columns=['ID hecho','gravedad','GRAVEDAD','camion','ciclista','transporte_publico','auto','moto','participantes', 'aaaa', 'mm', 'dd', 'fecha','hora','geocodificacion_CABA','VEHICULO_VICTIMA','direccion_normalizada','otra_direccion','calle','altura','cruce','tipo_calle','FECHA ','DD','MM'])
df_sin_columnas
```
