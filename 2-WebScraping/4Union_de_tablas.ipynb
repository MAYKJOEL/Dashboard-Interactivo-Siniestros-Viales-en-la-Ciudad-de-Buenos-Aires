En este proceso haremos la union de nuestra data Homicidios con los 3 webscraping que hemos realizado.

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import warnings
warnings.simplefilter("ignore")
```

Uniremos primero lo scrapeado con nuestra data homicidios:

* con las coordenadas hemos scrapeado los barrios
* con el numero de comunas hemos scrapeado sus nombres

```python
#Leemos el archivo reviews.parquet en un DataFrame
homicidiosdf = pq.read_table("../0-DATA/homicidiosETL.parquet").to_pandas()
barriosdf = pq.read_table("../0-DATA/BarriosBA.parquet").to_pandas()
comunasdf = pq.read_table("../0-DATA/comunasBA.parquet").to_pandas()
```

```python
homicidiosdf#.head(5)
```

```python
barriosdf.head(5)
```

```python
comunasdf.head(5)
```

uniendo estas datos

```python
df_merged = pd.merge(homicidiosdf, barriosdf, left_on=['Longitud','Latitud'],right_on=['pos x','pos y'], how='inner')
df_merged
```

```python
len(df_merged[df_merged.duplicated() == True])
```

```python
df_merged = df_merged.drop_duplicates()
```

```python
df_merged
```


Unimos nuevamente con el DF que faltaba

```python
df_merged2 = pd.merge(df_merged, comunasdf, left_on='Comunas',right_on='numero_comuna', how='inner')
df_merged2.head(5)
```

```python
len(df_merged2[df_merged2.duplicated() == True])
```

eliminando columnas iguales

```python
df_merged2 = df_merged2.drop(columns=["Comunas","numero_comuna",'pos x', 'pos y'])
df_merged2
```

Analizando los datos de las coordenadas

```python
df_merged2.Latitud.describe()
```

```python
df_merged2.Longitud.describe()
```

Como en ambos tenemos punto en ves de coordenadas, pór tanto eliminaremos aquellas filas donde no existe el nombre de Barrio.

```python
df_merged2 = df_merged2.dropna(subset=['Barrio'])
# Eliminar la primera fila
df_merged2 = df_merged2.drop(index=0)

# Restablecer los índices
df_merged2 = df_merged2.reset_index(drop=True)
df_merged2
```

```python
df_merged2.Latitud.describe()
```

```python
df_merged2.Longitud.describe()
```

Guardamos

```python
#guardamos en un CSV
df_merged2.to_csv("../0-DATA/homicidiosEDAunion.csv", index=False, encoding="utf-8") 
```
